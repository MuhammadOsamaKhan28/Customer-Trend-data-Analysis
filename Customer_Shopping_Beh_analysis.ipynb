```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/customer_shopping_behavior.csv")
df.head(3)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# df.isnull().sum()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# df['age'].min()
# df['age'].maxn()
```

```python
labels = ['young_Adult','Adult','Middle_aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
# Purchase column purchase_frrequancy_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' :7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' :14,
    'Every 3 Months' :90,
    'Annually' :365
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df['purchase_frequency_days'].value_counts()
```

```python
df = df.drop(columns=['promo_code_used'],axis=1)
```
